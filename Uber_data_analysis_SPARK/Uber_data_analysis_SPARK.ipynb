```python
from pyspark import SparkContext, SparkConf
```

```python
## Creating a spark Context : sc

sparkconf = SparkConf().setAppName("UBER").setMaster("local[4]")
sc = SparkContext(conf = sparkconf)
```

```python
# Loading from HDFS
data = sc.textFile("/uber.csv",2,use_unicode=True)
data.take(5)

# Loading from local

#from pandas import read_csv
#data = read_csv("/data/uber.csv") 
```

```python
type(data)
```

```python
## Removing Header

data_header = data.first()
data_no_header = data.filter(lambda x : x != data_header)
data_no_header.take(3)
```

```python
## Splitting the data

data_elements = data_no_header.map( lambda x: x.split(","))

data_elements.cache()
data_elements.take(3)
```

```python
data_elements.count()
```

```python
type(data_elements)
```

```python
data_elements.map( lambda x: x[0]).distinct().count()
```

```python
## Finding the number of distinct elements in dispatching_base_number ?

data_elements.map( lambda x: x[0]).distinct().collect()
```

```python
## Filter 'B02617'?

data_elements.filter( lambda x: 'B02617' in x[0]).count()
```

```python
## Sorting the elements (ascending)

data_elements.map( lambda x: (x[0], int(x[3]))).reduceByKey( lambda a,b: a + b).sortBy( lambda x: -x[1]).collect()
```
